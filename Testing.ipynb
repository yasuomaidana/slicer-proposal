```python
import numpy as np

arr = np.ones((4,4))*1/4

i4 = np.eye(4,4)

n1 = i4 - arr
n2 = 3/4*i4
print(n2)
tetra_v1 = np.array([[0.5,0.5,0.5,0],[0.5,0,0,0],[0,0.5,0,0],[1,1,0,0]])
tetra_v1
```

```python
print(tetra_v1)
print(n1@tetra_v1)
```

```python
p1 = np.array([0.5,0.5,0.5,0])
p2 = np.array([0.5,0,0,0])
p3 = np.array([0,0.5,0,0])
p4 = np.array([1,1,0,0])
tetra_v1 = np.vstack([p1,p2,p3,p4])
tetra_v2 = np.hstack((p1.reshape(4,1),p2.reshape(4,1),p3.reshape(4,1),p4.reshape(4,1)))
print(n1@tetra_v1)
# print(n1@tetra_v2)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial import Delaunay

fig= plt.figure()

points = tetra_v1[:,:-1]
tri= Delaunay(points)

ax = fig.add_subplot(projection='3d')

tr = tri.simplices[0]  # indices of first tetrahedron
pts = points[tr, :]  # pts is a 4x3 array of the tetrahedron coordinates

# plotting the six edges of the tetrahedron
for ij in [[0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]]:
    ax.plot3D(pts[ij, 0], pts[ij, 1], pts[ij, 2])
```

```python
fig= plt.figure()

points = n1@tetra_v1
points = points[:,:-1]
tri= Delaunay(points)

ax = fig.add_subplot(projection='3d')

tr = tri.simplices[0]  # indices of first tetrahedron
pts = points[tr, :]  # pts is a 4x3 array of the tetrahedron coordinates

# plotting the six edges of the tetrahedron
for ij in [[0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]]:
    ax.plot3D(pts[ij, 0], pts[ij, 1], pts[ij, 2], c='b')

points = tetra_v1[:,:-1]
tri= Delaunay(points)
tr = tri.simplices[0]  # indices of first tetrahedron
pts = points[tr, :]  # pts is a 4x3 array of the tetrahedron coordinates

for ij in [[0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]]:
    ax.plot3D(pts[ij, 0], pts[ij, 1], pts[ij, 2], c='k')
```
